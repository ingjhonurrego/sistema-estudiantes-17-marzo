##### Aca todo se muestra desde adentro pero no se pueden usar afuera

```python
def casa():
    p1 = input("Ingresa nombre: ")
    p2 = input("Ingresa nombre: ")
    p3 = input("Ingresa nombre: ")
    p4 = input("Ingresa nombre: ")
    
    print(p1,p2,p3,p4)
    
casa()
```

##### Mejor manera de mostrar datos: 

```python
def casa():
    p1 = input("Ingresa nombre: ")
    p2 = input("Ingresa nombre: ")
    p3 = input("Ingresa nombre: ")
    p4 = input("Ingresa nombre: ")
    
    return p1, p2, p3, p4
    
    
p1, p2, p3, p4 = casa()

print(p1)
print(p2)
print(p3)
print(p4)

```

##### Es de mejor manera hacerlo asi debido a que: 
##### Se separan responsabilidades 
##### La funcion -----> obtiene datos
##### Afuera ------> Se decide que hacer con ellos
